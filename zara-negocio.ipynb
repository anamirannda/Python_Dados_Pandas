```python
import pandas as pd

tabela_negocio = pd.read_csv('Zara_sales_EDA.csv',sep=';')
display(tabela_negocio)
```

```python
tabela_zara = tabela_negocio.dropna()
display(tabela_zara)
```

```python
tabelaZ = tabela_zara.groupby('season').agg({'Sales Volume':'sum','price':'mean'}).map('{:.1f}'.format)
display(tabelaZ)
```

```python
import matplotlib.pyplot as plt
tabela_zara.groupby('season').agg({'Sales Volume':'sum','price':'sum'}).plot(kind='bar')
plt.show()
```
